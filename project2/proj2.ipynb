## Setup

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from utils import vis_hybrid_image, load_image, save_image
from project2 import my_imfilter, create_hybrid_image
%load_ext autoreload
%autoreload 2

image1 = load_image('images/dog.bmp')
image2 = load_image('images/cat.bmp')

# display the dog and cat images
plt.figure(figsize=(3,3)); plt.imshow((image1*255).astype(np.uint8));
plt.figure(figsize=(3,3)); plt.imshow((image2*255).astype(np.uint8));
```

## Create filter

```python
cutoff_frequency = 7
filter = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,
                               sigma=cutoff_frequency)
filter = np.dot(filter, filter.T)

# let's take a look at the filter!
plt.figure(figsize=(4,4)); plt.imshow(filter);
```

## Apply filter to image
Both functions you need to implement in this project can be found in _student_\__code.py_. This code block will use your _my_\__imfilter_ function to create and display a blurry version of the dog image.

```python
blurry_dog = my_imfilter(image1, filter)
plt.figure(); plt.imshow((blurry_dog*255).astype(np.uint8));
save_image('../results/blurred_image.jpg', blurry_dog)
```

## Create hybrid image
Next, implement _create_\__hybrid_\__image_, which takes two images and makes a hybrid image using the low frequency content from one image and the high frequency content from another.

```python
low_frequencies, high_frequencies, hybrid_image = create_hybrid_image(image1, image2, filter)
vis = vis_hybrid_image(hybrid_image)
```

# Writeup

The aim of this project was to get familiar with using convolution as an image processing tool. In order to accomplish that, I implemented several functions that turn the theory into practice.

The support pillar of this project is my function convolve(channel, filter). This function takes a 2D array and applies a filter to it as described by the definition of convolution. It first determines which dimension of the image is larger and based on that, it pads the edges of the picture with zeroes in order to prevent the filter calculations from reaching an index out of bounds. Two simple for-cycles are used to go over all pixels in the image, but in order to speed up the process of multiplying the filter kernel with pixel values and adding it all together, the more efficient and parallelized NumPy functions were used. Specifically, NumPy.multiply and NumPy.sum were used. Their argument is always the range of pixels corresponding to the area covered by the filter kernel. Once the calculation is complete, the filtered channel is returned.

The function which utilizes this is the my_imfilter(image, filter). It uses functions from my previous project to split the image into separate channels, send them over to the convolve function and merge the filtered channels into the final image. The function works with provided greyscale images of Albert Einstein and Marilyn Monroe as well as the colored ones.

```python
plt.figure(figsize=(5, 5)); plt.imshow(load_image('../results/blurred_image.jpg'));
```

Finally, we have the create_hybrid_image(image1, image2, filter) function. It applies a low frequency filter on image1, subtracts this filter from image2 to create a high frequency version of the image, and finally merges the two images by adding them together. Once again, NumPy functions were used to speed up and simplify these mundane operations. The test runs seemed to work fine, all hybrid images resemble one of the images from up close and the other from further away, which was the desired goal.


```python
plt.figure(figsize=(5, 5)); plt.imshow(load_image('../results/hybrid_image.jpg'));
```
