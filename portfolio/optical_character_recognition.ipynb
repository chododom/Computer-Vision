# Optical character recognition (OCR)

For this project, I wanted to find a tool to perform OCR with and learn how to use it. I decided to use the Python version of the Tesseract OCR (pytesseract).

Tesseract uses artificial intelligence in order to interpret the text embedded in the image and does so in two main steps. The process consists of two steps. Firstly, Tesseract uses pixel patterns to recognize single characters. However, not all characters can be clearly recognized and that's where the second step comes into play, as it uses contextual semantics to figure out what the undetermined characters may be.

Paper on Tesseract OCR: https://static.googleusercontent.com/media/research.google.com/en//pubs/archive/33418.pdf
