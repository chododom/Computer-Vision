# Normalize Each Color Channel, rather than the Greyscale

```python
from src.project1 import *
%matplotlib inline

img = loadppm("images/zebra.ascii.ppm")
red = GetRedPixels(img)
green = GetGreenPixels(img)
blue = GetBluePixels(img)

normR = normalize(red)
normG = normalize(green)
normB = normalize(blue)

normIMG = mergeChannels(normR, normG, normB)

plt.imshow(normIMG)
plt.show()
```

# Create a greyscale conversion function that weights colors more realistically

Instead of calculating the greyscale value as a simple average where Greyscale = (R + G + B / 3), we use a different ratio of the colour channels. Since the three different colours have three different wavelengths and have their own contribution in the formation of images, we have to take average according to their contribution. Red colour has the longest wavelength and green colour is considered to be more soothing to the eyes, so we will increase the contribution of green, decrease red and fill the rest with blue. A good ratio (source: https://www.tutorialspoint.com/dip/grayscale_to_rgb_conversion.htm) is R: 30 %, G: 59 % and B: 11 %.

```python
from src.project1 import *
%matplotlib inline

zebra = loadppm("images/zebra.ascii.ppm")
greyArr = advancedGreyscale(zebra, 0.3, 0.59, 0.11)
plt.imshow(greyArr,cmap='gray', vmin=0, vmax=255)
plt.show()
```
