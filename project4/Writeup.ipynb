# Writeup

The aim of this assignment was to get familiar with how corner detection works in computer vision tasks. We used the Harris corner detector and applied non-local maxima suppression in order to reduce duplicity in strong corner markers. 

Harris Feature Detector:
Using a Sobel filter, derivatives were calculated for horizontal (Iy) and vertical (Ix) edges of the image. We then created a structure tensor for each pixel, which includes four different versions of the image (Ix * Ix, Iy * Iy, Ix * Iy, Iy * Ix), each convolved with a gaussian kernel. This tensor is then applied as the matrix for calculating an R value of the image using the formula R = Det(A) - k * Trace(A)^2 where k = 0.04. All this is done in the function HarrisDetector.
The function works fast for the simple testimage.pgm, where it runs for about 7 seconds, but for the nd1.jpg, it runs over a minute.

Non-local Maxima Suppression:
All pixels' R-values are sorted in decreasing order. The largest value (most significant feature) is saved as a global maximum and its suppression radius is set to infinity. As for the rest of the pixels, the minimum distance to points with higher values is saved as their suppression radius. List of points is sorted in decreasing order according to the points' suppression radii and given number of feature coordinates is returned

Visualization of the features (corners) is done using cv2's circle drawing function.
